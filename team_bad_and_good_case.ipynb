```python
from agno.team import Team
from agno.agent import Agent
from agno.models.google import Gemini

france = Agent(
    name="France Professional",
    role="France Professional",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 프랑스 전문가야. 항상 프랑스 관련 질문을 받으면 불어로 대답해.",
    instructions=[
        "프랑스와 관련된 질문이 나오면 항상 불어로 대답해.",
        "프랑스에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)


usa = Agent(
    name="USA Professional",
    role="USA Professional",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 미국 전문가야. 항상 프랑스 관련 질문을 받으면 영어로 대답해.",
    instructions=[
        "미국과 관련된 질문이 나오면 항상 영어로 대답해.",
        "미국에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)

korean = Agent(
    name="Korean Professional",
    role="Korean Professional",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 한국 전문가야. 항상 프랑스 관련 질문을 받으면 한국어로 대답해.",
    instructions=[
        "한국과 관련된 질문이 나오면 항상 한국로 대답해.",
        "한국에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)

team = Team(
    model=Gemini(id="gemini-flash-lite-latest"),
    members=[
      france,
      usa,
      korean,
    ],
    markdown=True
)

response = team.run("에펠탑에 대해서 설명해.")

print(response.content)
```

```python
from agno.team import Team
from agno.agent import Agent

france = Agent(
    name="France Professional",
    role="France Professional (Return in French)",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 프랑스 전문가야. 사용자의 언어에 상관없이 항상 프랑스 관련 질문을 받으면 불어로 대답해.",
    instructions=[
        "**프랑스와 관련된 질문이 나오면 사용자의 언어에 상관없이 항상 __불어__로 대답해.**",
        "프랑스에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)


usa = Agent(
    name="USA Professional",
    role="USA Professional (Return in English)",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 미국 전문가야. 사용자의 언어에 상관없이 항상 프랑스 관련 질문을 받으면 영어로 대답해.",
    instructions=[
        "**미국과 관련된 질문이 나오면 사용자의 언어에 상관없이 항상 __영어__로 대답해.**",
        "미국에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)

korean = Agent(
    name="South Korea Professional",
    role="South Korea Professional (Return in Korean)",
    model=Gemini(id="gemini-flash-lite-latest"),
    description="너는 한국 전문가야. 사용자의 언어에 상관없이 항상 프랑스 관련 질문을 받으면 한국어로 대답해.",
    instructions=[
        "**한국과 관련된 질문이 나오면 사용자의 언어에 상관없이 항상 __한국어__로 대답해.**",
        "한국에 대한 전문지식 및 역사를 기반으로 대답해."
    ]
)

team = Team(
    name="Multi-Countries Professional Team",
    model=Gemini(id="gemini-flash-lite-latest"),
    members=[
      france,
      usa,
      korean,
    ],
    instructions=[
        "You are a team of agents that can answer a variety of questions about each country.",
        "You can use your country-specific professional agents to answer the questions.",
        "NEVER DO NOT answer directly, you MUST orward the question to a member agent.",
        "Translate user question before forwarding it to each suitable member.",
    ],
    markdown=True
)

response = team.run("프랑스 에펠탑에 대해서 설명해.")

print(response.content)
```
