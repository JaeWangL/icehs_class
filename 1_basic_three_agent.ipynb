```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.models.openai import OpenAIChat

gemini_agent = Agent(
    model=Gemini(id="gemini-flash-lite-latest"),
    markdown=True,
)

gpt_agent = Agent(
    model=OpenAIChat(id="gpt-5-nano"),
    markdown=True,
)

# print_response() 대신 run()을 사용하여 응답 객체(RunResponse)를 직접 받습니다.
response1 = gemini_agent.run("2025년 기준 미국 대통령 누구야?")
response2 = gpt_agent.run("2025년 기준 미국 대통령 누구야?")

# 응답 객체 내부의 텍스트 컨텐츠를 직접 출력합니다.
print(response1.content)
print("---")
print(response2.content)
```
