```python
import os

# Google GenAI API Key 설정
os.environ["GOOGLE_API_KEY"] = "{{API_KEY}}"

print("Environment variables set.")
```

```python
from agno.agent import Agent
from agno.models.google import Gemini

agent = Agent(
    model=Gemini(id="gemini-flash-lite-latest"),
    markdown=True,
)

# print_response() 대신 run()을 사용하여 응답 객체(RunResponse)를 직접 받습니다.
response = agent.run("2025년 기준 미국 대통령 누구야?")

# 응답 객체 내부의 텍스트 컨텐츠를 직접 출력합니다.
print(response.content)
```

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools.googlesearch import GoogleSearchTools

google_search_tool = GoogleSearchTools()

agent = Agent(
    model=Gemini(id="gemini-flash-lite-latest"),
    tools=[google_search_tool],
    markdown=True,
    show_tool_calls=True  # 도구가 실행되는 과정을 로그로 표시
)

# 도구를 사용하여 검색 수행
response = agent.run("Agno Python 라이브러리의 최신 기능에 대해 검색해줘.")

print(response.content)
```

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools.duckduckgo import DuckDuckGoTools

duckduckgo_tool = DuckDuckGoTools()

agent = Agent(
    model=Gemini(id="gemini-flash-lite-latest"),
    tools=[duckduckgo_tool],
    description="You are a search agent that helps users find the most relevant information using duckduckgo.",
    instructions=[
        "Given a topic by the user, respond with the 3 most relevant search results about that topic.",
        "Search for 5 results and select the top 3 unique items.",
        "Search in both English and Korean.",
    ],
    markdown=True,
)

# 도구를 사용하여 검색 수행
response = agent.run("2025년 기준 미국 대통령 누구야?")

print(response.content)
```
